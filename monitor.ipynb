```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)         # initiate notebook for offline plot
```

```python
q_df = pd.read_csv('training_model/q.csv')
#score_df = pd.read_csv('training_model/.csv')

line_q_mean=go.Scatter(y=q_df['mean'], x=q_df['update'])
line_q_min=go.Scatter(y=q_df['min'], x=q_df['update'])
line_q_max=go.Scatter(y=q_df['max'], x=q_df['update'])
#line_score=go.Scatter(y=score_df['mean'])

data = [line_q_mean, line_q_min, line_q_max]

iplot(data)
```
