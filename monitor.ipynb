```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
from plotly import tools

init_notebook_mode(connected=True)         # initiate notebook for offline plot
```

```python
def print_q(start_pos=9000):
    q_df = pd.read_csv('training_model/q.csv')
    #score_df = pd.read_csv('training_model/.csv')

    q_df = q_df[start_pos:]
    line_q_mean=go.Scatter(y=q_df['mean'], x=q_df['Timestamp'], name='mean q')
    line_q_min=go.Scatter(y=q_df['min'], x=q_df['Timestamp'] , name='min q')
    line_q_max=go.Scatter(y=q_df['max'], x=q_df['Timestamp'] , name='max q')
    #line_score=go.Scatter(y=score_df['mean'])

    data = [line_q_mean, line_q_min, line_q_max]
    layout = go.Layout(title='q value')
    fig = go.Figure(data=data, layout=layout)

    iplot(fig)

print_q()
```

```python
start_pos = 10000
show_time = False

print_q(start_pos)

def print_score(start_pos):
    if show_time:
        x_label = 'Timestamp'
    else:
        x_label = 'run'
        
    score_df = pd.read_csv('training_model/score.csv')[start_pos:]
    line_score=go.Scatter(y=score_df['mean'], x=score_df[x_label] , name='Score')
    data = [line_score]
    
    layout = go.Layout(title='Score')
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
    
def print_step(start_pos):
    if show_time:
        x_label = 'Timestamp'
    else:
        x_label = 'run'
        
    step_df = pd.read_csv('training_model/step.csv')[start_pos:]
    line_step=go.Scatter(y=step_df['mean'], x=step_df[x_label] , name='Step',yaxis='y2')
    data = [line_step]
    
    layout = go.Layout(title='Step')
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)

        
def print_epsilon(start_pos):
    epsilon_df = pd.read_csv('training_model/epsilon.csv')[start_pos:]
    line_epsilon=go.Scatter(y=epsilon_df['mean'], x=epsilon_df['Timestamp'])
    data = [line_epsilon]
    
    layout = go.Layout(title='Epsilon')
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)

    
def print_loss(start_pos):
    loss_df = pd.read_csv('training_model/loss.csv')[start_pos:]
    line_loss=go.Scatter(y=loss_df['mean'], x=loss_df['Timestamp'])
    data = [line_loss]
    
    layout = go.Layout(title='Loss')
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
    
print_score(start_pos)
print_step(start_pos)
print_epsilon(start_pos)
print_loss(start_pos)



```
