```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
from plotly import tools

init_notebook_mode(connected=True)         # initiate notebook for offline plot
```

```python
start_pos = 0
show_time = False

def print_score(start_pos):
    df = pd.read_csv('training_model/score.csv')[start_pos:]
    print_result(start_pos, df, 'score',y_label='mean', x_label='run')
    
def print_step(start_pos):
    df = pd.read_csv('training_model/step.csv')[start_pos:]
    print_result(start_pos, df, 'step',y_label='mean', x_label='run')

def print_epsilon(start_pos):
    df = pd.read_csv('training_model/epsilon.csv')[start_pos:]
    print_result(start_pos, df, 'epsilon',y_label='mean', x_label='update')

    
def print_loss(start_pos):
    df = pd.read_csv('training_model/loss.csv')[start_pos:]
    print_result(start_pos, df, 'loss',y_label='mean', x_label='update')
     
def print_winning(start_pos):
    df = pd.read_csv('training_model/winning.csv')[start_pos:]
    print_result(start_pos, df, 'winning',y_label='current_mean', x_label='run')

def print_lossing(start_pos):
    df = pd.read_csv('training_model/lossing.csv')[start_pos:]
    print_result(start_pos, df, 'lossing',y_label='current_mean', x_label='run')

def print_wrong_move(start_pos):
    df = pd.read_csv('training_model/wrong_move.csv')[start_pos:]
    print_result(start_pos, df, 'wrong_move', y_label='current_mean', x_label='run')
    
def print_result(start_pos, df, label, y_label, x_label):
    if show_time:
        x_label = 'Timestamp'

        
    def prepare_df(df):
        
        df['current_mean'] = df['max'] - df['min']
        df['MA10'] = df['current_mean'].rolling(window=10).mean()
        df['MA30'] = df['current_mean'].rolling(window=30).mean()
        df['MA50'] = df['current_mean'].rolling(window=50).mean()
        df['MA100'] = df['current_mean'].rolling(window=100).mean()
    
        return df
    
    
    df = prepare_df(df)
    
    d1 =go.Scatter(y=df[y_label], x=df[x_label] , name=label)
    d2 =go.Scatter(y=df['MA10'], x=df[x_label] , name='MA10')
    d3 =go.Scatter(y=df['MA30'], x=df[x_label] , name='MA30')
    d4 =go.Scatter(y=df['MA50'], x=df[x_label] , name='MA50')
    d5 =go.Scatter(y=df['MA100'], x=df[x_label] , name='MA100')
    
    data = [d1,d2,d3,d4,d5]
    
    layout = go.Layout(title=label)
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)

    

    

def print_q(start_pos=9000):
    q_df = pd.read_csv('training_model/q.csv')
    #score_df = pd.read_csv('training_model/.csv')

    q_df = q_df[start_pos:]
    line_q_mean=go.Scatter(y=q_df['mean'], x=q_df['Timestamp'], name='mean q')
    line_q_min=go.Scatter(y=q_df['min'], x=q_df['Timestamp'] , name='min q')
    line_q_max=go.Scatter(y=q_df['max'], x=q_df['Timestamp'] , name='max q')
    #line_score=go.Scatter(y=score_df['mean'])

    data = [line_q_mean, line_q_min, line_q_max]
    layout = go.Layout(title='q value')
    fig = go.Figure(data=data, layout=layout)

    iplot(fig)

```
